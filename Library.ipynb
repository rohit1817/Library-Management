```python
import pandas as pd

df = pd.read_csv('/content/books.csv',index_col = False)
```

```python
df.head()
```

```python
df.info()
```

```python
df = df.drop('Unnamed: 12', axis=1)
```

```python
df.info()
```

```python
nan_values = df.isna()
any_missing_values = nan_values.any().any()
any_missing_values_in_column = nan_values.any()
missing_value_count = df.isnull().sum()
for column, has_missing in any_missing_values_in_column.items():
    if has_missing:
        count = missing_value_count[column]
        print(f" Column '{column}' has {count} missing values.")

print("\nMissing Values in the Entire DataFrame?")
print(any_missing_values)

print("\nMissing Values in Each Column?")
print(any_missing_values_in_column)

print("\nMissing Value Counts in Each Column:")
print(missing_value_count)
```

```python
df.dtypes
```

```python
df['average_rating'] = pd.to_numeric(df['average_rating'], errors='coerce')
df['isbn'] = pd.to_numeric(df['isbn'], errors='coerce')
df['isbn13'] = pd.to_numeric(df['isbn13'], errors='coerce')
df['  num_pages'] = pd.to_numeric(df['  num_pages'], errors='coerce')
df['publication_date'] = pd.to_datetime(df['publication_date'], errors='coerce')
error_count = df['publication_date'].isnull().sum()
df = df.dropna(subset=['publication_date'])
```

```python
df.info()
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
df.head()
```

```python
df.rename(columns={'  num_pages':'num_pages'})
```

```python

df['authors'] = df['authors'].str.replace('/', ' and ')
df['title'] = df['title'].str.replace('/',' and ')

#df['track_name'] = df['track_name'].str.replace(r'\(.*\)', '', regex=True)
df['authors'] = df['authors'].str.replace('[^a-zA-Z\s]', '', regex=True)
```
